```python
#DATA PRE-PROCESSING#

#1-Replace Missing Values with No Response#
df.fillna('No Response', inplace=True)

#2-Change the question titles into Themes#
df.rename(columns=

  {'What is your current job title or position?': 'Job Title',

             'Which industry do you work in?': 'Industry',

            'How many years of experience do you have in supply chain operations?':
            'Years Of Experience',

            'What is the size of your organization?':
            'Organization Size',

            'What are the most significant challenges you face in your supply chain operations in Egypt?':
            'Challenges in Egyptian Supply Chains',

            'How do external factors (e.g., political instability, economic fluctuations) affect your supply chain processes?':
            'Effects of external factors on supply chains',

            'What specific challenges do you encounter in inventory management and meeting customer demand?':
            'Inventory and Demand Challenges',

            'Does your organization currently use data analytics tools in supply chain operations? If yes, how are they used?':
            'Data Analytics Adoption',

            'What challenges do you face in integrating data from suppliers, distributors, or customers?':
            'Data Integration Challenges',

            'What are the biggest barriers to adopting advanced technologies like AI or PDA in your supply chain?':
            'Barriers to AI and PDA Adoption',

            "In your opinion, how could AI or PDA address challenges in Egypt's supply chain operations?":
            'Benefits of PDA Adoption in Egyptian Supply Chains',

            'What types of training or support would you need to adopt and use AI or PDA effectively?':
            'Training/Support Needs for AI/PDA',

            'What concerns do you have about implementing AI or PDA in your supply chain?':'AI/PDA Implementation Concerns'
             }, inplace= True)

#3-Group job positions#
def job_title_group(row):
    title = str(row).lower()
    if 'purchasing' in title or 'purchase' in title:
        return 'Purchasing'
    elif 'operations' in title:
        return 'Operations'
    elif 'logistics' in title:
        return 'Logistics'
    elif 'sales' in title:
        return 'Sales'
    elif 'Credit' in title:
        return 'Sales'
    else:
        return row

# Apply the function to the 'Job Title' column
df['Job Title'] = df['Job Title'].apply(job_title_group)

print(df['Job Title'])
```

```python
df.head(2)
```

```python
#Preprocessing Responses seperated by commas
import re

split_columns = ['Challenges in Egyptian Supply Chains','Effects of external factors on supply chains',
    'Inventory and Demand Challenges', 'Data Analytics Adoption',
    'Data Integration Challenges', 'Barriers to AI and PDA Adoption',
    'Benefits of PDA Adoption in Egyptian Supply Chains', 'Training/Support Needs for AI/PDA',
    'AI/PDA Implementation Concerns']

# Define a function to split while preserving commas inside parentheses
def smart_split(text):
    return re.split(r',\s*(?![^()]*\))', text) if pd.notna(text) else text

df_melted = df.melt(
    id_vars=[col for col in df.columns if col not in split_columns],
    value_vars=split_columns,
    var_name='Category',
    value_name='Response'
)

df_melted['Response'] = df_melted['Response'].apply(smart_split)
df_exploded = df_melted.explode('Response')
df_exploded.reset_index(drop=True, inplace=True)




print(df_exploded.head())


```
